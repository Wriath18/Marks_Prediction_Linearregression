```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv("./Training Data/Linear_X_Train.csv")
Y = pd.read_csv("./Training Data/Linear_Y_Train.csv")
```

```python
X_data = X.values
Y_data = Y.values
plt.scatter(X,Y)
```

```python
mean_x = X_data.mean()
mean_y = Y_data.mean()
```

```python
deviations_x = np.zeros((len(X_data),1))
devations_y = np.zeros((len(Y_data), 1))
devations_y
deviations_x
```

```python
for i in range(len(X_data)):
    deviations_x[i] = X_data[i] - mean_x
    devations_y[i] = Y_data[i] - mean_y

deviations_x
devations_y
```

```python
product_devations = deviations_x*devations_y
product_devations
```

```python
sum_product_deviations = np.sum(product_devations)
sum_product_deviations
```

```python
Square_devations_x = np.array([x**2 for x in deviations_x])
Square_devations_x
```

```python
m = sum_product_deviations / np.sum(Square_devations_x)
c = mean_y - (m * mean_x)
```

```python
plt.scatter(X, Y, label='Data points')
plt.plot(X, m*X + c, color='red', label='Best-fitting line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Best-fitting Line')
plt.legend()
plt.show()
```

```python
test_data = pd.read_csv("./Test Cases/Linear_X_Test.csv")
test_data = test_data.values
```

```python
y_predictions = np.zeros((len(test_data), 1))

for i in range(len(test_data)):
    y_predictions[i] = (m*test_data[i] + c)

y_predictions
plt.scatter(test_data, y_predictions)
```

```python
predictions_data = pd.DataFrame(y_predictions)
predictions_data.to_csv('predictions.csv', index=False)
```
